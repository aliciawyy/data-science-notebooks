```python
n = 100
np.random.seed(0)
pi = stats.bernoulli(0.4)
delta = pi.rvs(n)
g1 = np.random.normal(4, 0.9, n)
g2 = np.random.normal(1, 0.5, n)

y = (1. - delta) * g1 + delta * g2
```
